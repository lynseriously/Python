# Scrapying Pictures - IMDB TOP 250

```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
defaultpath = '/Users/lichen/Desktop'
path = os.path.join(defaultpath,'Pics')
if not os.path.exists(path):
    try:
        os.makedirs(path)
    except OSError as exception:
        if exception.errno != errno.EEXIST:
            raise
```

```python
url = r'https://www.imdb.com/chart/top/'
r = requests.get(url)
soup = BeautifulSoup(r.content,'lxml')
```

```python
sectionlinks = soup.find_all("td", class_ = 'posterColumn')
i = 0
namelist = []
piclist = []
for sectionlink in sectionlinks:
    i += 1
    namelist.append(str(i)+". "+ sectionlink.find_all('img')[0].get('alt'))
    piclist.append(sectionlink.find_all('img')[0].get('src'))
    
i = 0
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1985.125 Safari/537.36',
         'Referer': str(url)}

for piclink in piclist:
    picname = path+'/'+namelist[i] + '.jpg'

    result = requests.get(piclink, headers = headers )

    imgfile = open(picname,'wb')
    imgfile.write(result.content)
    imgfile.close()
    i += 1
```

# Scrapying Text - From Dynamic JavaScript Website

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service as ChromeService
```

```python
#beautifulsoup cant handle javascript enabled website, thus use selenium
options = Options()
#options.add_argument("--headless")
options.add_argument("window-size=1500*1500")
chrome_service = ChromeService(r'/Users/lichen/Desktop/chromedriver.exe')
driver = webdriver.Chrome(service = chrome_service, options = options)
params = {'behavior': 'allow', 'downloadPath': '/Users/lichen/Desktop'}
driver.execute_cdp_cmd('Page.setDownloadBehavior', params)
driver.get(r'https://finviz.com/news.ashx')

soup = BeautifulSoup(driver.page_source,'html.parser')
driver.quit()
```

```python
newslist = []
news = soup.find_all("a", class_ = 'nn-tab-link')
for i, title in enumerate(news):
    newslist.append(str(i) + ". " + title.get_text())
newslist
```
